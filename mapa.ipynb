```python
import folium
from shapely.geometry import Point
import geopandas as gpd

# Coordenadas proporcionadas
x, y = -58.47533969, -34.68757022

# Crea un punto con shapely
point = Point(x, y)

# Crea un GeoDataFrame con el punto y define el sistema de referencia espacial (CRS)
gdf = gpd.GeoDataFrame(geometry=[point], crs="EPSG:4326")

# Crea un mapa de folium centrado en las coordenadas
m = folium.Map(location=[y, x], zoom_start=15)

# AÃ±ade el GeoDataFrame como una capa al mapa
folium.GeoJson(gdf).add_to(m)

# Muestra el mapa interactivo
m
```
