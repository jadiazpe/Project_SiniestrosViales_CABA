### >>> Cargue de fuentes de datos

```python
df_victimas = pd.read_csv('./Datasets/ETL/victimas.csv')        # ---> DataFrame obtenido del primer trabajo de ETL via PowerBI  
df_hechos = pd.read_csv('./Datasets/ETL/hechos.csv')            # ---> DataFrame obtenido del primer trabajo de ETL via PowerBI
```

### 2. Preparación del DataFrame 'df_hechos'

```python
df_hechos.head()
```

```python
df_hechos.info()
```

```python
# ---> Remover carácteres no deseados de la columnas: POSX and POSY
df_hechos[['POSX', 'POSY']] = df_hechos[['POSX', 'POSY']].map(lambda x: str(x).replace("'", ""))
```

```python
df_hechos.head()
```

### x. MERGING DataFrames 'df_victimas' and 'df_hechos' for EDA 

```python
# ---> MERGING los DataFrames: df_victimas and df_hechos
df_sv = pd.merge(df_victimas, df_hechos, on='ID', how='left')
df_sv.head()
```

```python
df_sv.info()
```

```python
df_sv.duplicated().sum()                          # ---> Number of duplicate rows
```

```python
# ---> Sending CSV clean file for EDA processing
df_sv.to_csv('./Datasets/EDA/victimas_hechos.csv', encoding='utf-8', index=False)
```
